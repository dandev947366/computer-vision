```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#image folder path
img_path = '/Users/dandev947366/Desktop/Projects-github/computer-vision/computer-vision/img/01/1.JPG'
```

```python
img = Image.open(img_path)
plt.imshow(img)
```

```python
im_gray = img.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(img)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Histogram')
plt.xlabel('Pixel values')
plt.ylabel('Quantity')
```

```python
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image") 
plt.show()
```
