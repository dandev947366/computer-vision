Rotate images with Transpose

```python
from PIL import Image
from imgtools import * 
from IPython.display import display # display images
```

```python
#image file path
my_img_path = '/Users/dandev947366/Desktop/Projects-github/computer-vision/computer-vision/img/01/1.jpg'
```

```python
img = load_image(my_img_path)

display(img)
```

```python
# rotate image
rotated_image = img.rotate(90)
display(rotated_image)
```

```python
# Transpose image
transposed_image = img.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```
