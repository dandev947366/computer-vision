Read single image and its info

```python
from PIL import Image
from scipy import linalg, ndimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
#image folder path
image_path = '/Users/dandev947366/Desktop/Projects-github/computer-vision/computer-vision/img/01/animal.jpg'

#read image
img01 = Image.open(image_path)

gray_image = img01.convert('L')
image_array = np.array(gray_image)
threshold = 223
binary_image = np.where(image_array > threshold, 0, 1)

labeled_image, num_features = ndimage.label(binary_image)

fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Gray Image')

plt.subplot(3,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')

plt.subplot(3,1,3)
plt.imshow(labeled_image, cmap='gray')
plt.title(f'Labeled Image ({num_features} objects)')

plt.show()
```
