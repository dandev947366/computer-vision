Using Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'bo-', label="Data")
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label="Data2")
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color='g', alpha=0.6)
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Pie chart")
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Data')
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
data = [15,  18, 22, 30, 35, 45, 50 , 55, 65]
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

```python
data = [15,  18, 22, 30, 35, 45, 50 , 55, 65]
plt.violinplot(data)
plt.title('Violin Plot')
plt.ylabel('Values')
plt.show()
```

```python
from wordcloud import WordCloud

text = 'Python is an amazing language for data visualization and word clouds'

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()
```
