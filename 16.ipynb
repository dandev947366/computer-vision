```python
from PIL import Image
from scipy import linalg, ndimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
#image folder path
image_path = '/Users/dandev947366/Desktop/Projects-github/computer-vision/computer-vision/img/01/img_avg_1.JPG'
image = Image.open(image_path)
image.show()
```

```python
gray_image = image.convert('L')
image_a = np.array(gray_image)

gradient_x =ndimage.sobel(image_a, axis=0, mode='constant')
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient Image')
```

```python
gray_image = image.convert('L')
image_a = np.array(gray_image)

gradient_y =ndimage.sobel(image_a, axis=1, mode='constant')
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.imshow(gradient_y, cmap='gray')
plt.title('Gradient Image')
```
