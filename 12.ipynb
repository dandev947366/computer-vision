```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/Users/dandev947366/Desktop/Projects-github/computer-vision/computer-vision/img/01/1.JPG'
```

```python
im = Image.open(image_path)
```

```python
im_array = np.array(im)
print(im_array.shape)

im_array_2 = np.array(im.convert('L'), 'f')
print(im_array_2.shape)
print(im_array_2.dtype)
```

```python
print(im_array_2.shape)
inverted_image = 255 - im_array_2
squared_image = im_array_2 ** 2
clamped_image = np.clip(im_array_2, 100, 200)
# print(inverted_image)
# print(squared_image)
# print(clamped_image)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(im_array_2, cmap='gray')
axs[0, 0].set_title('Original image')

axs[0, 1].imshow(inverted_image, cmap='gray')
axs[0, 1].set_title('Inverted image')

axs[1, 0].imshow(squared_image, cmap='gray')
axs[1, 0].set_title('Squared image')

axs[1, 1].imshow(clamped_image, cmap='gray')
axs[1, 1].set_title('Clamped image')

plt.show()

```
